```python
import numpy as np
```

```python
## forward mode
vneg = 0; v0 = 0; v1 = vneg**2; v2 = 1-vneg; v3 = v2**2
v4 = v0 - v1; v5 = v4**2; v6 = 100*v5; v7 = v6 + v3

## for x1
dvnegx1 = 1; dv0x1 = 0; dv1x1 = 2*vneg*dvnegx1; dv2x1 = -1*dvnegx1; dv3x1 = 2*v2*dv2x1
dv4x1 = dv0x1 - dv1x1; dv5x1 = 2*v4*dv4x1; dv6x1 = 100*dv5x1; dv7x1 = dv6x1+ dv3x1
dydx1_fwd = dv7x1

## for x2
dvnegx2 = 0; dv0x2 = 1; dv1x2 = 2*vneg*dvnegx2; dv2x2 = -1*dvnegx2; dv3x2 = 2*v2*dv2x2
dv4x2 = dv0x2 - dv1x2; dv5x2 = 2*v4*dv4x2; dv6x2 = 100*dv5x2; dv7x2 = dv6x2 + dv3x2
dydx2_fwd = dv7x2
```

```python
## reverse mode
dydv7 = 1

## v7 = v6 + v3
dydv6 = dydv7 * 1 ; dydv3 = dydv7 * 1

## v6 = 100*v5
dydv5 = dydv6 * 100

## v5 = v4**2
dydv4 = dydv5 * 2 * v4

## v3 = v2**2
dydv2 = dydv3*2*v2

## v4 = v0 - v1
dydv0 = dydv4; dydv1 = dydv4*(-1)

## v2 = 1 - vneg; v1 = vneg**2
dydvneg = dydv2*(-1) + dydv1*(-2*vneg)

dydx1_rev = dydvneg
dydx2_rev = dydv0
```

```python
print(dydx1_fwd, dydx2_fwd)
print(dydx1_rev, dydx2_rev)
```

```python
## pytorch autograd
import torch
x1 = torch.tensor(0.0, requires_grad=True)
x2 = torch.tensor(0.0, requires_grad=True)

y = 100 * (x2 - x1**2)**2 + (1 - x1)**2
```

```python
y.backward()
```

```python
print(x1.grad.item(), x2.grad.item())
```
